```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('IceCreamData (1).csv')
df
```

```python
resultado = df.sort_values('Temperature', ascending=  False)
resultado
```

```python
celsius = (resultado['Temperature']-32*5/9)
resultado['Temperature']= celsius.round(3)
resultado = resultado.rename(columns={'Revenue': 'Receita em dólar'})
resultado = resultado.rename(columns={'Temperature': 'Temperatura'})
resultado
```

```python
plt.plot(resultado['Temperatura'],
         resultado['Receita em dólar'],
         'r--',
         label='US$')
```
